```python
# Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load and read dataset
data = pd.read_csv('/content/poverty.csv')
data.head()
```

```python
#determine the location of the dataset 
data=data.iloc[4:]
```

```python
#location of the year data set
data_year=data.iloc[:,4:]
```

```python
#read the data of the year
data_year.head()
```

```python
#poverty data per year
data_year_poverty=pd.DataFrame(data_year)
data_year_poverty
```

```python
#sum of the years for better results
data_year=data_year.sum(axis=0)
data_year
```

```python
#plot the year poverty graph 
plt.figure(figsize=(10,10))
plt.plot(data_year_poverty)
plt.title('Poverty per year')
plt.xlabel('Povery rate')
plt.ylabel('Povery rate')
```

```python
#data type change
data_year_poverty['year']=data_year_poverty.index.astype('int')
data_year_poverty
```

```python
#read the data type
data_year_poverty.dtypes
```
